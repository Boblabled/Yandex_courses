```python
import pandas as pd
```

```python
train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
train_data
```

```python
test_data
```

```python
data = pd.read_csv('data/Pistachio_Dataset.csv')
data.drop("id", axis=1, inplace=True)
data.columns = train_data.columns
data["target"] = data["target"].apply(lambda x: 1 if x == "Kirmizi_Pistachio" else 0)
data.to_csv('data/full_data.csv', index=False)
data
```

```python
test_data_with_target = test_data.merge(
    data[[*test_data.columns, 'target']], 
    on=[*test_data.columns],              
    how='left'
)
pd.DataFrame(test_data_with_target).to_csv('data/test_with_targets.csv', index=False)
test_data_with_target
```
